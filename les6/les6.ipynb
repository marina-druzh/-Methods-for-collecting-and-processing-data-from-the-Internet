# 5. Проверка, что записалось в Mongo

```python
from pymongo import MongoClient
client = MongoClient('127.0.0.1', 27017)
# print(client.list_database_names())
db = client['db_castorama']
print(db.list_collection_names())
```

# Запись в json

```python
import json
from bson import ObjectId


class create_dict(dict): 
  
    # __init__ function 
    def __init__(self): 
        self = dict() 
          
    # Function to add key:value 
    def add(self, key, value): 
        self[key] = value 
        
mydict = create_dict()
i = 1
for mycol in db.list_collection_names():
  i = 1
  for y in db[mycol].find():
    mydict.add(i, ({'id': str(y['_id']), "images":y['images'],"link":y['link'],"name":y['name'], "price": y['price']}))
    i = i+1
  f_name = f'{mycol}_json'
  print(f_name)
  with open(f_name, 'w', encoding='UTF-8') as f:
          json.dump(mydict, f)
```

```python
with open('castorama_ru_json', 'r', encoding='UTF-8') as f:
    ff = json.load(f)
print(f'количество товаров в файле: {len(ff)}\n')
```
